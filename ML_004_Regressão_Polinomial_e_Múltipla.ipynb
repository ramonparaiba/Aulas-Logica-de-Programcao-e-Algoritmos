#Regressão polinomial

Regressão polinomial é uma técnica estatística que busca modelar a relação entre uma variável dependente e uma ou mais variáveis independentes através de uma equação polinomial. Ao contrário da regressão linear, a regressão polinomial permite ajustar curvas mais complexas e não lineares aos dados. O objetivo é encontrar a equação que melhor descreve a relação entre as variáveis e permitir a previsão de valores da variável dependente com base nas variáveis independentes. A regressão polinomial é amplamente utilizada em muitas áreas, incluindo engenharia, física, economia e ciências sociais, entre outras.

---



```python
#importações
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Velocidade do Carro x Hora do dia
hora = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
velocidade = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
```

```python
plt.scatter(hora, velocidade)
plt.show()
```

Desenhar a linha da regressão Polinomial

```python
modelo = np.poly1d(np.polyfit(hora, velocidade, 3))
linha = np.linspace(1, 22, 100)

plt.scatter(hora, velocidade)
plt.plot(linha, modelo(linha))
plt.show()
```

```python
#r-quadrado - varia de 0 - 1. Sendo 0 nenhuma relação e 1 100% relacionado
from sklearn.metrics import r2_score

print(r2_score(velocidade, modelo(hora)))
```

Previsão de valores futuros

```python
#Qual será a velocidade de um carro que transita em horas diferentes do dia?
#Para tal, precisamos do modelo já criado modelo = np.poly1d(np.polyfit(hora, velocidade, 3))

print(modelo(18))
print(modelo(18.5))
print(modelo(12))
print(modelo(10))
```

Ajuste Ruim

```python
x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

modelo = np.poly1d(np.polyfit(x, y, 3))
linha = np.linspace(2, 95, 100)

plt.scatter(x, y)
plt.plot(linha, modelo(linha))
plt.show()
```

```python
# Pela análise do gráfico, suspeito que o r-quadrado deve dar baixo

print(r2_score(y, modelo(x)))
```

#Regressão Múltipla
É um método estatístico utilizado para avaliar a relação entre uma variável dependente (variável resposta) e duas ou mais variáveis independentes (variáveis explicativas). É uma extensão da regressão linear simples, que utiliza apenas uma variável independente para prever a variável dependente.

```python
#Importação do pandas
import pandas as pd
```

```python
#importar base de dados

data = pd.read_csv('/content/drive/MyDrive/datasets-colab/data.csv')
```

```python
#showdata
data
```

```python
#Criar data frames com as seções que serão as variáveis independentes
X = data[['Weight','Volume']]
#Criar dataframe com a variável dependente
y = data['CO2']

#É comum nomear a lista de valores independentes com X maiúsculo
#E nomear a lista de valores dependentes com y minúsculo
```

```python
#
from sklearn import linear_model
```

```python
#Criar um objeto de regressão linear
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
#O Objeto está pronto para prever os valores de emissão de CO2 com base no peso e no volume do motor de um carro
predictCO2 = regr.predict([[2300, 1300]])
print(predictCO2)
```

```python
#Coeficiente da regressão
print(regr.coef_)
#Peso = 0.00755095 ; A cada Kg aumentado, a emissão de CO2 aumenta a 0.00755095g
#volume = 0.00780526 ; a cada cm3 aumentado, a emissão de CO2 aumenta em 0.00780526g
```

```python
predictCO2 = regr.predict([[3300, 1300]])
print(predictCO2)
```
